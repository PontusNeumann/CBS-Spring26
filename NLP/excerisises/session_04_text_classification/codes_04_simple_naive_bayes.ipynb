# <span style="color:blue;">Simple Sentiment Analysis with Naïve Bayes</span>



## <span style="color:brown;">Step 1: Data (Three Sentences with Labels)</span>
| Sentence              | Sentiment |
|----------------------|----------|
| "I love NFTs"       | Positive (+) |
| "NFTs are bad"      | Negative (-) |
| "NFTs are amazing"  | Positive (+) |

---

## <span style="color:brown;">Step 2: Preprocessing (Lowercase, Tokenize, Remove Stopwords)</span>
| Original Sentence      | Preprocessed Tokens |
|-----------------------|---------------------|
| "I love NFTs"        | `["love", "nfts"]` |
| "NFTs are bad"       | `["nfts", "bad"]` |
| "NFTs are amazing"   | `["nfts", "amazing"]` |

---

## <span style="color:brown;">Step 3: Feature Extraction (Word Presence)</span>
**Feature Words Set:**  
`{"love", "nfts", "bad", "amazing"}`  

| Sentence Tokens       | Features (Word Presence) | Sentiment |
|----------------------|-----------------------------------------------------|----------|
| `["love", "nfts"]`   | `{'love': True, 'nfts': True, 'bad': False, 'amazing': False}` | Positive |
| `["nfts", "bad"]`    | `{'love': False, 'nfts': True, 'bad': True, 'amazing': False}` | Negative |
| `["nfts", "amazing"]`| `{'love': False, 'nfts': True, 'bad': False, 'amazing': True}` | Positive |

---


## <span style="color:brown;">Step 4: Train a Naïve Bayes Classifier</span>
- **Counts word occurrences** in positive and negative examples.
- **Calculates probabilities** of each word appearing in each class.
- **Applies Bayes’ theorem** to classify new sentences.

---

## <span style="color:brown;">Step 5: Predict Sentiment of a New Sentence</span>
| New Sentence      | Preprocessed Tokens | Feature Representation |
|------------------|--------------------|-------------------------------------------------|
| "nfts are good" | `["nfts", "good"]` | `{'love': False, 'nfts': True, 'bad': False, 'amazing': False}` |

Prediction:Positive (+)
