## <font color='blue'> Walkthrough Example - Sentiment Analysis using TextBlob and VADER <font>

    
    
### <font color='blue'>Example 1: Extracting Sentiment Scores for Words (VADER Sentiment Lexicon)<font>

 https://github.com/cjhutto/vaderSentiment/blob/master/vaderSentiment/vader_lexicon.txt

```python
#!python3 -m pip install vaderSentiment
#!python3 -m pip install --upgrade pip

# download the VADER lexicon from NLTK's data repository
#nltk.download('vader_lexicon')

import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# OR from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Access the lexicon directly through the SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Define a function to check sentiment score of a word
def get_word_senti(word):
    return analyzer.lexicon.get(word.lower(), "Word not found in VADER lexicon.")

# Test words
words = ["The", "previous", "lecture", "was", "BORING","I", "like", "Sentiment", "Analysis", "lecture"]

for word in words:
    print(f"{word} → Sentiment Score: {get_word_senti(word)}")
```

## <font color='blue'> Example 2: Sentiment Analysis on a sample review (VADER and TextBlob) <font>


```python
sample_review = ["The previous lecture was BORING!", "I like Sentiment Analysis lecture"]
print("Sample Review: \n", sample_review, "\n")

# Using TextBlob
from textblob import TextBlob

# Analyzing each review separately
for review in sample_review:
    post = TextBlob(review) # For each review, creating a TextBlob object 
    tsa = post.sentiment
    print(f"TextBlob Result: '{review}'\n", tsa) #Returns named tuple with Polarity and Subjectivity
    print(f"  Polarity: {round(tsa.polarity, 2)}, Subjectivity: {round(tsa.subjectivity, 2)}\n")

vdr = SentimentIntensityAnalyzer()

# Analyzing each review separately
for review in sample_review:
    vsa = vdr.polarity_scores(review) #returns a dictionary containing the sentiment scores for the review
    print(f"\n VADER Result: '{review}'\n", vsa)
    print(f"  Positive: {round(vsa['pos'], 2)}, Neutral: {round(vsa['neu'], 2)}, "
          f"Negative: {round(vsa['neg'], 2)}, Compound: {round(vsa['compound'], 2)}\n")

```

  ##  Sentiment Scoring                             

<font color='blue'> For Textblob (polarity -1 (very negative) to 1 (very positive)) <font>

Positive: Polarity score > 0
    
Neutral: Polarity score = 0
    
Negative: Polarity score < 0
                            
<font color='blue'> The subjectivity is from 0.0 (very objective->factual) to 1.0 (very subjective->opinion based).
                            
                            
<font color='blue'> For VADER (compound: Aggregated polarity score from -1 (very negative) to 1 (very positive)) <font>
    
 Positive: Compound score > 0.05
    
Neutral: Compound score between -0.05 and 0.05
    
Negative: Compound score < -0.05 


## <font color='blue'> Example-2 Sentiment Analysis on a large Dataset (csv) <font>
    
URL to download data:  https://www.kaggle.com/datasets/andrewmvd/trip-advisor-hotel-reviews
    
    
### <font color='black'> Convert scores to sentiment category <font>
    

<font color='blue'> In the given Dataset ( rating are from 1-5) <font>
    
 Positive: rating >= 4
    
Neutral: rating =3
    
Negative: rating <=2
                                

                            

```python
import pandas as pd

#sets the maximum width of the display to 1000 characters
pd.set_option('display.width', 1000)

# Load the dataset
file_path = 'tripadvisor_hotel_reviews.csv'
data = pd.read_csv(file_path)

# Print the first few rows of the dataset to understand its structure
print(data.head(5))
```

##  <font color='blue'> Sentiment Analysis with TextBlob <font>

```python
# import library
from textblob import TextBlob

# Create a DataFrame (with two additional columns) to store the results
t_dframe = pd.DataFrame(columns=['Review', 'Polarity', 'Subjectivity'])

t_dframe['Review'] = data['Review'].head(1000)

# Adding columns for TextBlob results
#dframe['Polarity'] = 0.0
#dframe['Subjectivity'] = 0.0

# Iterate through the first 1000 reviews
for j, rev in enumerate(t_dframe['Review']):
    # Creates a TextBlob object
    txb = TextBlob(rev)
    # Analyze sentiment and store the sentiment analysis results
    t_dframe.loc[j, 'Polarity'] = round(txb.sentiment.polarity,2)
    t_dframe.loc[j, 'Subjectivity'] = round(txb.sentiment.subjectivity, 2)

# Print the first 10 rows of the DataFrame
print(t_dframe.head(10))


```

##  <font color='blue'> Sentiment Analysis with VADER <font>

```python

# Initialize the SentimentIntensityAnalyzer
vsa = SentimentIntensityAnalyzer()

# Create a DataFrame to store the results
v_dframe = pd.DataFrame(columns=['Review', 'Compound', 'Positive', 'Negative', 'Neutral'])


# Alternate dframe = pd.DataFrame()
v_dframe['Review']=data['Review'].head(1000)

# Iterate through the first 1000 reviews in the 'data' DataFrame
for i, review in enumerate(v_dframe['Review']):
    # Analyze sentiment using VADER
    vader_result = vsa.polarity_scores(review)         
    v_dframe.loc[i, 'Compound'] = round(vader_result['compound'],2)
    v_dframe.loc[i, 'Positive'] = round(vader_result['pos'],2)
    v_dframe.loc[i, 'Negative'] = round(vader_result['neg'],2)
    v_dframe.loc[i, 'Neutral'] = round(vader_result['neu'],2)
#.loc function-used to access a specific row and column in the DataFrame.
# Allow to specify the row by its index (i) and the column by its name ('Positive')

# Print the first 10 rows of the DataFrame
print(v_dframe.head(20))
```
