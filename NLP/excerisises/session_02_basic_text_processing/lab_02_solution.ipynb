### Åroblem (1) Information Extraction using Regular Expression

(1.1) Create and print a list of three to four employees containing the information in the following format:

    Employee Name: Email address: Telephone Number

(1.2) Use regular expressions in Python to extract information from the above list and display in the following format:

    Employee Name:  Email address: Employee Department

(1.3) Use regular expressions in Python to extract the name and telephone information from the list and display in the following format:

    Employee Name: Telephone Number without country code

<font color="blue"> Hint: You can see some code examples about using regular expressions in Python.<font>
https://regexone.com/references/python


```python
#(1.1)

Employee_Record = ['Rajani Singh: rs.digi@cbs.dk: +918750602050',
                   'Sine Zambach: sz.itm@cbs.dk: +4538153815',
                    'Daniel Hardt: dha.msc@cbs.dk: +41852024']
print('1.1 Created List\n''Employee Name:',' Email address:',' Telephone Number')
for d in Employee_Record:
    print(d)

#(1.2)
#\w is short hand for [a-zA-Z_]
#\d is short hand for [0-9]

import re
my_pattern =r'(\w+\s\w+): (\w+\.(\w+)@\w+\.\w+)'
print('\n1.2 Extracted Information \nEmployee Name:',' Email address:','Department')
for d in Employee_Record:
    match = re.search(my_pattern ,d)
    print(match.group(1)+': ' + match.group(2) + ': ' + match.group(3))

#(1.3)
#In tel. number we select last 8 digits, after the prefix (which is set to optional using ? sign)

new_pattern =r'(\w+\s\w+): \w+\.\w+@\w+\.\w+: \+?\d?\d?(\d{8,10}$)'
print('\n1.3 Extracted Information \nEmployee Name:','Telephone Number')
for d in Employee_Record:
    match = re.search(new_pattern ,d)
    print(match.group(1)+': ' + match.group(2))

    #notice all the match groups - and how they work - in the first one we have 3, in the second one only two - always use them for the info you want to retrieve
```

match group: used for capturing groups

g1:name

g2:email

g3:is nested in the email and takes the domain

groups are denoted by ()

### Problem (2) Tokenization

 (2.1) Create a sample document consist of 4-5 sentences.
    
 (2.2) Split the sample document into sentences and word tokens and print them.

 (2.3) Use Prettytable to print them in nice tabular format.
    
 (2.4) Do the sentences and word tokenization using NLTK, print them in pretty table format.Compare both the results.

 (2.5) Compare both the results of NLTK and Textblob tokenization.

```python
#(2.1)
from textblob import TextBlob
import nltk
nltk.download('punkt')

sample_doc = """
            Denmark is a Nordic country in Northern Europe.
            Copenhagen, its capital, is known for colorful houses and the iconic Little Mermaid statue.
            Denmark is famous for its strong social welfare system and high standard of living.
            """

print('\n (2.1) Sample Document:\n', sample_doc)

#(2.2)

sample_doc= sample_doc.replace("\n\t", "") # remove extra space and newlines
my_text = TextBlob(sample_doc) #instantiate TextBlob on sample_doc
s_token=my_text.sentences # Tokenize by sentences
w_token=my_text.words # Tokenize by words


print("\n (2.2)  Sentence Tokenization:\n ", s_token)
print("\n (2.2) Word Tokenization: \n", w_token,"\n")

#(2.3)

from prettytable import PrettyTable

#Create PrettyTable with 2 columns, and specify the column names
sen_table = PrettyTable(['Index','Sentence as Tokens'])
i=0
print("\n(2.3) Tokenization in Textblob\n")

#iterate over sentences in table and append them to new rows of the table
for t in s_token:
    sen_table.add_row([i,t])
    i += 1

print(sen_table)

word_table = PrettyTable(['Index','Word as Tokens '])
i=0
#iterate over sentences in table and append them to new rows of the table
for t in w_token:
    word_table.add_row([i,t])
    i += 1
print(word_table)


# (2.4)
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from prettytable import PrettyTable

#Create PrettyTable with 2 columns
s_table = PrettyTable(['Index','Sentence as Tokens'])
i=0

print("\n(2.4) Tokenization in NLTK\n")
#iterate over sentences in table and append them to new rows of the table
for t in sent_tokenize(sample_doc):
    s_table.add_row([i,t])
    i += 1
print(s_table)

#Create PrettyTable with 2 columns
w_table = PrettyTable(['Index','Word as Tokens '])
i=0
#iterate over sentences in table and append them to new rows of the table
for t in word_tokenize(sample_doc):
    w_table.add_row([i,t])
    i += 1
print(w_table)
```

Notice how in textblob we have to import the whole class and instantiate and then use the methods

In nltk we only import the function and can immediately use it

In nltk we are also including punctuation in tokenization

### Problem (4) Part of Speech (POS) Tagging

(4.1) Use NLTK library to do POS tagging, check if the word is stopword and print them in the following format:  [Word    | POS (NLTK)  |  Is a Stop word? ]
    
(4.2) Use spaCy library for task (4.1) and print as: [Word    | POS (spacy)  |  Is a Stop word? ]
    
(4.3) Compare both the results.

<font color="brown"> Hint: You can look into this blog for more information. https://towardsdatascience.com/named-entity-recognition-with-nltk-and-spacy-8c4a7d88e7da <font>

```python
# POS tag list:

# CC	coordinating conjunction
# CD	cardinal digit
# DT	determiner
# EX	existential there (like: "there is" ... think of it like "there exists")
# FW	foreign word
# IN	preposition/subordinating conjunction
# JJ	adjective	'big'
# JJR	adjective, comparative	'bigger'
# JJS	adjective, superlative	'biggest'
# LS	list marker	1)
# MD	modal	could, will
# NN	noun, singular 'desk'
# NNS	noun plural	'desks'
# NNP	proper noun, singular	'Harrison'
# NNPS	proper noun, plural	'Americans'
# PDT	predeterminer	'all the kids'
# POS	possessive ending	parent\'s
# PRP	personal pronoun	I, he, she
# PRP$	possessive pronoun	my, his, hers
# RB	adverb	very, silently,
# RBR	adverb, comparative	better
# RBS	adverb, superlative	best
# RP	particle	give up
# TO	to	go 'to' the store.
# UH	interjection	errrrrrrrm
# VB	verb, base form	take
# VBD	verb, past tense	took
# VBG	verb, gerund/present participle	taking
# VBN	verb, past participle	taken
# VBP	verb, sing. present, non-3d	take
# VBZ	verb, 3rd person sing. present	takes
# WDT	wh-determiner	which
# WP	wh-pronoun	who, what
# WP$	possessive wh-pronoun	whose
# WRB	wh-abverb	where, when

#(4.1) Part-of-speech tagging in NLTK
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
nltk.download('punkt')
from prettytable import PrettyTable
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

stop_words = set(stopwords.words('english'))

sample_text = """Denmark is a Nordic country in Northern Europe.
            Copenhagen, its capital, is known for colorful houses and the iconic Little Mermaid statue.
            Denmark is famous for its strong social welfare system and high standard of living.
            """
print('\n (4.1) Part-of-speech tagging in NLTK:\n')

words = nltk.word_tokenize(sample_text)
pos_tagged = nltk.pos_tag(words)
#print(pos_tagged)

table = PrettyTable(['Word', 'POS (NLTK)','Is a Stop word?'])
i=0
for w in words:
    if w in stop_words:
        check=True
    else:
        check=False
    table.add_row([w,pos_tagged[i][1],check])
    i+=1

print(table)

#(4.2) Part-of-speech tagging in spaCy
import spacy
from spacy.lang.en.examples import sentences
#instantiate spacy en language module
ok = spacy.load("en_core_web_sm")
doc=ok(sample_text)
print('\n (4.2) Part-of-speech tagging in spaCy:\n')
tab = PrettyTable(['Word', 'POS (spaCy)', 'Is a Stop word?'])
for w in doc:
    tab.add_row([w.text, w.pos_, w.is_stop])
print(tab)
```

### Problem (5) Named Entity Recognition

 (5.0) Create a sample text document (Ex: you can use email about an event or a conference invitation).
    
 (5.1) Define a function to do the Entity recontion by using NLTK library and print them as: [Entity | Label]

 (5.2) Do the same by using spaCy library. Visualize the results using displacy.
    
 (5.3) Compare both the results of NLTK and spaCy.
    
   <font color="brown"> Hint: You can look into this blog for more information. https://towardsdatascience.com/named-entity-recognition-with-nltk-and-spacy-8c4a7d88e7da <font>

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')

#(5.0)

sample_text = """Dear colleagues,

We are happy to inform you that the Economics Department at the University of Crete will host the 12th Oligo Workshop, from June 6 to June 8, 2024.
The Workshop, which is organized annually by the Oligo Society, is privileged this year to have as keynote speakers E. Auriol (Toulouse School of Economics), V. Denicolo (University of Bologna) and P. Heidhues (DICE).

Submission deadline: March 15, 2024

Workshop Webpage: https://oligoworkshop2024.soc.uoc.gr/
Inquiries and questions: oligoworkshop2024@gmail.com

Looking forward to seeing you in Crete.
"""


import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from prettytable import PrettyTable

#sample_text = "This is a sample text containing named entities like Apple and New York."

sentences = sent_tokenize(sample_text)
print('\n(5.1) NER with NLTK')
tab1 = PrettyTable(['Entity', 'Label'])

def NER_nltk():
    try:
        for s in sentences:
            words = nltk.word_tokenize(s)
            tagged = nltk.pos_tag(words)
            namedEnt = nltk.ne_chunk(tagged)
            #namedEnt.draw()

            for e in namedEnt:
                try:
                    tab1.add_row([''.join(c[0] for c in e.leaves()), e.label()])
                except AttributeError:
                    pass
        print(tab1)
    except Exception as e: #caughts any errors
        print(str(e))

NER_nltk()



#(5.2) NER with spaCy
import spacy
from spacy import displacy
#instantiate spacy en language module
lm = spacy.load("en_core_web_sm")
#displacy.render(lm(str(doc))), jupyter=True, style='ent')
print('\n (5.2) NER with spaCy')

#function takes a doc (string) and prints NER table with all identified entities from the doc
def NER_spacy (doc):
    tab2 = PrettyTable(['Entity', 'Label'])
    #obtain NER labels from document entities using .label_ method
    #NER classes: 	Location, Person, Organization, Money, Percent, Date, Time
    for e in doc.ents:
        tab2.add_row([e.text, e.label_])

    print(tab2)

#run the language module on the sample_text
my_doc = lm(sample_text)
NER_spacy(my_doc)
displacy.render((my_doc), jupyter=True, style='ent') #this is for the colour NEW
displacy.render((my_doc.sents), style='dep', jupyter = True,options = {'distance': 150}) #this is for the dependency

```
