### Counting Types and Tokens in a corpus

```python
import nltk
from nltk.tokenize import word_tokenize

# Replace 'corpus.txt' with the file path of your corpus 
#with open('/Users/jenny/nltk_data/corpora/abc/science.txt', 'r', encoding='utf-8') as file:
 #   corpus_text = file.read()
    
from nltk.corpus import abc
corpus_text=abc.raw('science.txt')
# Tokenize the corpus text
tokens = word_tokenize(corpus_text)

# Calculate the number of tokens and types
total_tokens = len(tokens)
total_types = len(set(tokens))

print(" Total number of tokens:", total_tokens)
print("Total number of types:", total_types)
```

## Removal of Stopwords 

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords if not already downloaded
#nltk.download('stopwords')
#nltk.download('punkt')

# Tokenize the text
#text="The stars twinkled in the dark, illuminating the night sky."
text="NLTK and spaCy are popular NLP libraries used for text processing."
tokens = word_tokenize(text)

# Load English stopwords
en_stop = set(stopwords.words('english'))
#print(en_stop)
# Remove stopwords
new_tokens = [word for word in tokens if word.lower() not in en_stop]

print("NLTK stopwords removal: ",new_tokens)

#Stopwords removal in spacy
import spacy

# Load English model
nlp = spacy.load('en_core_web_sm')

#mytext="The stars twinkled in the dark, illuminating the night sky."
mytext="NLTK and spaCy are popular NLP libraries used for text processing."
doc = nlp(mytext)

# Remove stopwords
filtered_tokens = [t.text for t in doc if not t.is_stop]

print("spaCy stopwords removal: ",filtered_tokens)
```

## Stemming


```python
from nltk.stem.snowball import SnowballStemmer
#from nltk.tokenize import word_tokenize
from nltk.tokenize import TreebankWordTokenizer

sbs = SnowballStemmer("english", True)
text="The stars twinkled in the dark, illuminating the night sky."
method=TreebankWordTokenizer()
words =method.tokenize(text)
stemmed_words = [sbs.stem(w) for w in words]
#print("Original:", text)
#print("Tokenized:", words)
#print("Stemmed:", stemmed_words)

#Print in prettytable
from prettytable import PrettyTable
table = PrettyTable(['Word','Stemmed'])
for w in words:
    table.add_row([w,sbs.stem(w)])
print(table)
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import TreebankWordTokenizer
text="The stars twinkled in the dark, illuminating the night sky."
method=TreebankWordTokenizer()
words =method.tokenize(text)
lemma = WordNetLemmatizer()

#Create PrettyTable for word lemma

from prettytable import PrettyTable
  
table = PrettyTable(['Word','Word Lemma'])
for w in words:
    table.add_row([w,lemma.lemmatize(w)])
print(table)
```
