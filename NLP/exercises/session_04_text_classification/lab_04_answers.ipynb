### Problem 2: Data Pre-processing
    
Create a function 'data_preprocess(text)' in NLTK that takes the text data as input and returns the pre-processed word tokens as 'new_text'. Inside the function apply the following pre-processing techniques:  
    
(2.1) Convert input text into lowercase.
    
(2.2) Tokenize the text into word tokens.
    
(2.3) Remove stopwords.
    
(2.4) Apply stemming or lemmatization into word tokens.
    
Preview the data to ensure pre-processing is as expected.

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download stopwords from NLTK
#nltk.download('punkt')
#nltk.download('stopwords')

def data_preprocess(text):
    # (2.1) lowercase the text
    text = text.lower()     
    # (2.2) Tokenize the text into words
    tokens = word_tokenize(text)
    new_tokens=[]
    # (2.3)  Remove stopwords
    ps = PorterStemmer()
    for w in tokens:
        if w not in stopwords.words('english'):
    # (2.4) Stem the words
            stem_tokens = ps.stem(w)   
            new_tokens.append(stem_tokens)  # Add the stemmed word to new_tokens
    return  new_tokens

# Apply preprocessing to each message
d_frame['message'] = d_frame['message'].apply(data_preprocess)

# Preview the data to ensure pre-processing is as expected
print(d_frame.head(10))
```

### <font color='white'> Problem 3: Feature Extraction and data split: <font>
    
(3.1) Convert text data into numerical features by using the Bag of Words. Use the top 1000 words as your features.
    
 (3.2) Split the dataset into train (75%) and test (25%) sets.

```python
#(3.1)

# Convert DataFrame rows into a list of tuples, each containing message and label
ml_list = [(m,l) for m,l in zip(d_frame['message'], d_frame['label'])]

# Create a list of all word tokens
w_list = []
for w in d_frame['message']:
    w_list.extend(w)

# Create a frequency distribution of all words
w_count = nltk.FreqDist(w_list)

# Use the top 1000 words as features
top_words = list(w_count.keys())[:1000]

# Define a Feature Extraction Function
def extract_features(doc):
    #convert the doc into a set of unique words
    w_set = set(doc)
    # initializes an empty dictionary
    f_dict = {}
    for w in top_words:
        # checks whether that word is present in the words set. 
        # The result is a Boolean value (True if the word is present, False otherwise)
        f_dict[w] = True if w in w_set else False  
    return f_dict

# Created Feature Sets
f_sets = [(extract_features(m),l) for (m,l) in ml_list]

# (3.2) Split 75% of the data for training, 25% for testing
size = int(len(f_sets)*0.75)
train_set, test_set = f_sets[:size], f_sets[size:]
```

###  Problem 4: Naive Bayes Classifier Training
    
(4.1) Train a Naive Bayes classifier using the features extracted in Problem 3. 
 
(4.2) Show most informative 20 features. 
    
(4.3) Predict the labels for the test set.

```python
from nltk import NaiveBayesClassifier # For creating and using a Naive Bayes classifier
from collections import defaultdict # For dictionaries with default values for missing keys

# (4.1) Train the Naive Bayes Classifier
nb = NaiveBayesClassifier.train(train_set)

#(4.2) Most informative features
nb.show_most_informative_features(20)

# (4.3) Get the list of actual class from the test set and predicted class from the classifier
predicted_label=[]
actual_label=[]
for f, l in test_set:
    predicted_label.append(nb.classify(f)) 
    actual_label.append(l)
```

### Problem 5: Model Evaluation:
         
(5.1) Calculate the accuracy of the model.
     
(5.2) Calculate precision, recall, F1-score for both classes (spam and ham) and print them in table format.

```python
import nltk.metrics
from nltk.metrics import precision, recall, f_measure, accuracy
from collections import defaultdict


# (5.1) Accuracy
a = round(accuracy(actual_label, predicted_label),2)
print(f"Accuracy of the Naive Bayes classifier is {a*100}%. \n")


#(5.2) Precision, Recall, F-Measure

# key -> label 
# refsets value -> set of indices representing the test instances that truly belong to that class.
# checksets: value -> set of indices of test instances that were predicted to belong to the class.
refsets = defaultdict(set)
checksets = defaultdict(set)

for i, (f, l) in enumerate(test_set):
    refsets[l].add(i)
    observed = nb.classify(f)
    checksets[observed].add(i)

precision_spam = round(precision(refsets['spam'], checksets['spam']),2)
recall_spam = round(recall(refsets['spam'], checksets['spam']),2)
f_measure_spam = round(f_measure(refsets['spam'], checksets['spam']),2)

# Precision, Recall, F-Measure for neg classes

precision_ham = round(precision(refsets['ham'], checksets['ham']),2)
recall_ham = round(recall(refsets['ham'], checksets['ham']),2)
f_measure_ham = round(f_measure(refsets['ham'], checksets['ham']),2)

#  Print in table format using Panda library <font>

row_labels = ['Spam', 'Ham']

table = {
    ' Precision(%)': [precision_spam*100, precision_ham*100],
    ' Recall(%)': [recall_spam*100, recall_ham*100],
    ' F1-Measure(%)': [f_measure_spam*100, f_measure_ham*100]
}
df = pd.DataFrame(table, index=row_labels)
print(df)
```
