## <font color='blue'> Unzip the data folder <font>

```python
#Download the data as zip file 'nft.zip' and use the code below to unzip it into 'nft' folder.

from zipfile import ZipFile #import ZipFile class from the zipfile module
import os

print (os.getcwd()) 

# Unzip the uploaded dataset
z_path = "archive.zip"
uz_folder = "nft"

# Extract the zip file
with ZipFile(z_path, 'r') as my_file: #opens the ZIP file (z_path) in read mode ('r')
    my_file.extractall(uz_folder)

# List the files in the unzipped directory to understand its structure
unzipped_files = os.listdir(uz_folder)
#unzipped_files
```

## <font color='blue'> Reading a sample csv file <font>

```python
import pandas as pd

# Load one of the CSV files to examine its structure
#file_path = os.path.join(uz_folder, unzipped_files[0])
file_path="nft/240209_SOL_1777.csv"
# Read data from CSV files into a DataFrame by using pd.read_csv() function.
sample = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
sample.head(6)
```

## <font color='blue'> Creating one single meta data file <font>

```python
# Load all CSV files and append them into a single dataframe

# Initialize an empty list to store the DataFrames
all_data_frames = []

# Iterate over each file in unzipped_files
for f in unzipped_files:
    if f.endswith('.csv'): # Check if the file ends with '.csv'
        one_path = os.path.join(uz_folder, f)  # Construct the full path to the CSV file
        df = pd.read_csv(one_path)             # Read the CSV file into a DataFrame
        all_data_frames.append(df)             # Append the DataFrame to the list

# Concatenate all DataFrames into a single DataFrame
full_data = pd.concat(all_data_frames, ignore_index=True)
print("Single data file named 'full_data' has been created sucessfully!")


full_data.head(12)
```
