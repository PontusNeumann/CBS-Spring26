 ### <font color=brown> 1.1 Download a corpus  <font>

```python
# make sure that NLTK and reuters corpus is accessible
# Error! "Resource reuters not found" - try nltk.download('reuters')  to download

# Open Finder → press Cmd + Shift + G → go to or ~/ and Search for nltk_data
import nltk
from nltk.corpus import abc

print('Total number of fileids in a corpus: ',len(abc.fileids()),'\n')
print('Name of fileids in a corpus: ',abc.fileids(),'\n')
```

 ### <font color=brown> 1.2 Count the words in a corpus <font>

```python
from collections import Counter
counts = Counter(abc.words('science.txt'))   #print(counts)
total_count = len(abc.words('science.txt'))
print('Total numer of words in a corpus: ', total_count,'\n')
print('Top 10 words with their counts: ',counts.most_common(10),'\n')
```

 ### <font color=brown> 1.3  Compute the frequencies of words <font>

```python
for w in counts:
    counts[w] = round(counts[w]/float(total_count),3)
counts 
#OR counts[word] /= float(total_count)
#print( sum(counts.values()))  # The probabilities should add up to 1
```

 ### <font color=brown> 1.4  Unigram model: Generate 100 words of language using frequencies of words  <font>

```python

import random
#from nltk import FreqDist 
gen_text = []
n=100
for _ in range(n):
    r = random.random()
    i = .0
    for w, c in counts.items():
        i += c
        if i >= r:
            gen_text.append(w)
            break    
print('Generated 100 words: \n',' '.join(gen_text))
print('\n ')

p=1.0
for w in gen_text:
    p=p*counts[w]
print('Probability of random generated text: ',p)

#Alternate:
#from operator import mul
#from functools import reduce
#print(reduce(mul, [counts[w] for w in text], 1.0)) 

```
