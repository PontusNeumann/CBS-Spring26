# Example - Multinomial Logistic Regression

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Sample data

sample_data =[
    ("The course effectively make complex NLP concepts understandable.", 'positive'),
    ("Assignments were aligned with the lecture content.",'neutral'),
    ("Some parts of the course felt too fast.",'negative'),
    ("Feedback was welcomed to improve the course.",'neutral'),
    ("The stuff we had to read was sometimes hard to understand.",'negative'),
    ("The course materials were up-to-date and quite relevant.",'positive'),
    ("Some lectures were overly technical without clear explanations.",'negative'),
    ("Hands-on projects enhanced my understanding of NLP.",'positive'),
    ("The course required prerequisites that were not listed.",'negative'),
    ("The online class system worked smoothly.",'neutral'),
    ("Teacher knew a lot and made learning fun.",'positive'),
    ("Not enough examples of real-world NLP applications.",'negative'),
    ("Doing exercises helped us understand NLP better.",'positive'),
    ("The course pace was consistent throughout the semester.",'neutral'),
    ("Very good NLP course.",'positive'),
    ("More guest speakers would have added value.",'neutral'),
    ("Talking with others in class was helpful.",'positive'),
    ("Grading was sometimes confusing.",'negative'),
    ("The course lasted for a good amount of time.",'neutral')
]


# Feature extraction and labels
comments, labels = zip(*sample_data)

# Vectorize comments by using Bag of words approach 
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(comments)

#split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.20, random_state=0)

# Train logistic regression model
# Info. about different solver: link https://medium.com/@arnavr/scikit-learn-solvers-explained-780a17bc322d
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train, y_train)

# Test the model for new comment
# "The course was very helpful for learning NLP."
# "The course was too technical."
# "The course duration seemed appropriate for covering the material."

new_comment = ["The course duration seemed appropriate for covering the material."]
new_features = vectorizer.transform(new_comment)
pred_label = model.predict(new_features)

print(f" Input comment: {new_comment[0]}\n Predicted Sentiment: {pred_label[0]}")
```
