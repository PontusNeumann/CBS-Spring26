# Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Create a figure and axis
fig, ax = plt.subplots()

# Set the title and labels
ax.set_title('Sigmoid Function')
ax.set_xlabel('x')
ax.set_ylabel('sigmoid(x)')

# Create a line object with no data
line, = ax.plot(x, sigmoid(x), 'b-', linewidth=2)
```

```python
# Calculate Sigmoid for the heart disease 3 features example

import numpy as np

# Estimated parameters
w0, w1, w2, w3 = -1.5, 0.04, 0.03, 0.02

# Patient data
age, bp, chol = 50, 130, 240

# Calculate z
z = w0 + w1 * age + w2 * bp + w3 * chol

# Calculate probability of having heart disease
P_Y1 = 1 / (1 + np.exp(-z))
print(P_Y1)
```
