## Part-of-speech (POS) tagging in NLTK

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from prettytable import PrettyTable
sample_text = "This is the second lecture of NLP course"
#sample_text="The cat sat on the mat"

words=nltk.word_tokenize(sample_text)
pos_tagged = nltk.pos_tag(words)
table = PrettyTable(['Word', 'POS Tag (NLTK)'])
i=0
for w in words:
    table.add_row([w,pos_tagged[i][1]])
    i+=1
print(table)
```

## Part-of-speech (POS) tagging in spaCy

```python
import spacy
# Load the 'en_core_web_sm' model
model = spacy.load("en_core_web_sm")
from prettytable import PrettyTable

sample_text = "This is the second lecture of NLP course"
doc=model(sample_text)
tab = PrettyTable(['Word', 'POS Tag (spaCy)'])
for w in doc:
    tab.add_row([w.text, w.pos_])
print(tab)
```

## Named Entity Recognition (NER) in NLTK


```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
from prettytable import PrettyTable
sentence = "Apple Saket opened on today in the heart of India’s thriving capital of New Delhi."
words = word_tokenize(sentence)
tagged = pos_tag(words)
namedEnt = nltk.ne_chunk(tagged)
#print(namedEnt)
table = PrettyTable(['Entity', 'Label (NLTK)'])
for e in namedEnt:
    try:
        table.add_row([''.join (c[0] for c in e),e.label()])
    except AttributeError:
        pass
print(table)
namedEnt.draw()
```

## Named Entity Recognition (NER) in spaCy


```python
import spacy
from prettytable import PrettyTable
model = spacy.load("en_core_web_sm")
#sentence="In New York City, Elon Musk's company Tesla plans to launch a car next Tuesday."
sentence = "Apple Saket opened on today in the heart of India’s thriving capital of New Delhi."
doc = model(sentence)
tab = PrettyTable(['Entity', 'Label (spaCy)'])
for e in doc.ents:
    tab.add_row([e.text, e.label_]) 
print(tab)

from spacy import displacy
displacy.render((doc), jupyter=True, style='ent')
displacy.render((doc.sents), style='dep', jupyter = True,options = {'distance': 150})
```
