```python
import nltk
from nltk.tokenize import sent_tokenize
import re

#text=['puma','Puma','Pumas','pumas']
text="""The puma, also known as the mountain lion or cougar,is a magnificent and solitary big cat native to the Americas. Pumas possess muscular bodies and sleek golden coats, making them formidable hunters in their habitats. Pumas primarily hunt deer and other ungulates,using their sharp claws and powerful jaws to deliver swift and lethal attacks. Encounters with pumas in the wild are rare due to their elusive nature and preference for remote areas. However, human-puma conflicts can occur when their habitats overlap with human settlements. Pumas play a vital role in maintaining the balance of their ecosystems as apex predators, regulating prey populations and contributing to biodiversity. Puma, an apex predator of the Americas, embodies strength, agility, and stealth in the wild."""
pattern = r"[pP]uma|[pP]anther|[cC]atamount |[mM]ountain lion|[cC]ougar"
sentences=sent_tokenize(text)
print(sentences,'\n')
for t in sentences:
    #check=re.search(pattern,t)
    #check=re.match(pattern,t)
    check=re.findall(pattern,t)
    print(check)
```
