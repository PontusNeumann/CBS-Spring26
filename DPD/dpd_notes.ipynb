## Lecture 2 — Competition in newly vulnerable markets (news & information services)

### 1) Why some markets become “newly vulnerable”

A **newly vulnerable market** is one that becomes:

* **Newly easy to enter** (digital distribution lowers entry barriers)
* **Attractive to attack** (attackers can “pick off” the best customers)
* **Difficult to defend** (incumbents lose control over distribution + attention) 

**Metaphor:** think of the market as a castle that used to have one main gate (print distribution / direct relationships). Digital platforms create *many side doors*, so attackers don’t need to storm the gate—they just walk in through the easiest door. 

---

### 2) The competitive battleground: attention, distribution, and data

In **news and information services** (e.g., travel advice like TripAdvisor), competition often shifts from “who has the best content” to **who controls the user journey**:

* Platforms re-route users to their own surfaces (search, feeds, maps) → **disintermediation** of the original service.
* They leverage **data-driven learning** to improve relevance faster than traditional firms. 

**Example (travel info):** If search results answer “best hotels in Copenhagen” directly, users may never reach a travel-advice platform—market share moves upstream to the distributor. 

---

### 3) Platforms and network effects (the core mechanism)

Many of these services are **two-sided platforms** with:

* **Same-side effects** (more users on *your* side changes value)
* **Cross-side effects** (more users on the *other* side changes value)
  …and these can be **positive or negative**. 

**Quick examples**

* **Positive cross-side:** more readers → more advertisers → more revenue → better content.
* **Negative same-side:** too many advertisers → **ad clutter** → worse user experience. 

---

### 4) Positive feedback and “tipping” (winner-take-all risk)

**Positive feedback** can create a **flywheel**: more users → more value → more users. If strong enough, markets can **tip** toward one dominant player (winner-take-all tendencies). 

What makes tipping more likely?

* **High intensity of network effects**
* Strong **complements** (apps, creators, reviews, partners) around the core product
* Users spending more time → more interaction opportunities → more value 

---

### 5) Switching costs and lock-in (why users don’t leave)

**Switching costs** aren’t just money—often they’re:

* habits, saved preferences, reputation history, social graph, workflows.
  This creates a **lock-in cycle** where leaving feels like “starting over.” 

**Key nuance:**
High market share ≠ high switching costs, and low market share ≠ low lock-in. You can have a small niche with intense lock-in. 

**Metaphor:** lock-in is like **Velcro**—each tiny hook is minor, but together it’s hard to pull away.

---

### 6) Data network effects (learning effects)

**Data network effects**: the platform gets better as it learns from usage data:

* more users → more data → better ML → better experience → more users (a learning loop). 

**Example (news/info):** Better personalization increases time spent, which produces more signals, which improves ranking/recommendations faster than a rival with less scale. 

---

### 7) Strategy lens shift: from industry analysis to network analysis

Lecture takeaway: strategy needs to shift toward **network logic**, focusing on:

* **Value creation dynamics:** network effects + learning effects, complements, clusters/synergies
* **Value capture dynamics:** pricing, **multi-homing**, disintermediation, network bridging 

**Metaphor:** competing isn’t only about building a better “product”—it’s about building a stronger **ecosystem web** that keeps pulling users back in.

## Lecturer Important Statements

* **Why are platforms like Snapchat struggling relative to TikTok?**
  In digital platform markets, strong **network effects**, **data-driven learning effects**, and high user engagement can create *positive feedback loops*. This can make it seem like “winner-takes-all,” where one platform captures a disproportionate share of attention and market power.

* **Can a company simply choose a market and scale it into existence? (No.)**
  Firms cannot arbitrarily “activate” an economy at scale. Market success depends on **demand conditions, user adoption, ecosystem participation, and network effects**—not just strategic intent.

* **The market is not necessarily demanding variety.**
  In some digital markets, users prefer **standardized, dominant platforms** because of convenience, coordination benefits, and network value—reducing the effective demand for fragmented alternatives.

* **The Volt case: Optimization drives conversion.**
  Continuous **data optimization and experimentation** (e.g., A/B testing, UX refinement, funnel improvements) can significantly increase conversion rates. Competitive advantage often comes from systematic optimization rather than radical innovation.

* **Using AI without differentiation wastes time.**
  AI tools can increase efficiency, but if firms rely on generic outputs, products become homogeneous. Sustainable advantage requires identifying a unique **edge**—proprietary data, brand, positioning, or ecosystem integration. Otherwise, AI may accelerate commoditization rather than differentiation.

* AI tools are chalanging traditional platforms by interaction and customer participation.

* GAFA(M), google, apple, facebook, amazon, microsoft

* The definition of **inertia** is the resistance to change, which can be due to switching costs, habits, or lack of alternatives. Inertia can create lock-in even without high switching costs.
