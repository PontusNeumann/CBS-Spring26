## Write a k-fold cross validation example using python using given CSV file which has four features. Fit LinearRegression model and evaluated for each fold. Calculate the Mean Squared Error (MSE) for each fold and outputs the average MSE.

```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('patient_visits.csv')
```

```python
# Define features and target variable
features = ['Month', 'Year', 'Patient_Spend', 'Patient_Visits']
target = 'Patient_Spend'  # Assuming we're predicting 'Patient_Spend'
```

```python
# Independent variables (X) and target variable (y)
X = data[['Month', 'Year', 'Patient_Visits']]  # Excluding the target from features
y = data[target]
```

```python
# Initialize k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# Initialize model
model = LinearRegression()
```

```python
# Track cross-validation results
fold = 1
mse_scores = []
```

```python
for train_index, test_index in kf.split(X):
    # Split data into train and test sets
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict on test set
    y_pred = model.predict(X_test)
    
    # Calculate Mean Squared Error
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)
    
    print(f"Fold {fold}: Mean Squared Error = {mse:.4f}")
    fold += 1
```

```python
# Average MSE across folds
average_mse = sum(mse_scores) / len(mse_scores)
print(f"\nAverage Mean Squared Error across all folds: {average_mse:.4f}")
```

## Demo - 02: Showcase the importance of validation test on output.csv file which has four features features = ['Month', 'Year', 'Patient_Spend', 'Patient_Visits'].

- Without Validation: Evaluate the model on the same data it was trained on.
- With Validation: Split the data into training and test sets, evaluate the model on unseen data (the test set).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('patient_visits.csv')
```

```python
# Define features and target variable
features = ['Month', 'Year', 'Patient_Spend', 'Patient_Visits']
target = 'Patient_Spend'  # Assuming we're predicting 'Patient_Spend'
```

```python
# Independent variables (X) and target variable (y)
X = data[['Month', 'Year', 'Patient_Visits']]  # Excluding the target from features
y = data[target]
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# # TODO: Initialize LinearRegression model
model = LinearRegression()
```

#### Case 1 - No Validation: The model is trained and evaluated on the same data (X_train and y_train), leading to overly optimistic result

```python
### Case 1: Training and Evaluating on the Same Data
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
train_mse = mean_squared_error(y_train, y_train_pred)
print(f"Case 1 - Without Validation (Same Data): Mean Squared Error = {train_mse:.4f}")
```

#### Case 2 - With Validation: The data is split into training and test sets using train_test_split. The model is trained on the training data (X_train, y_train) and evaluated on unseen test data (X_test, y_test), providing a realistic measure of performance.

```python
### Case 2: Training on Training Set, Evaluating on Test Set
model.fit(X_train, y_train)  # Train on training data
y_test_pred = model.predict(X_test)  # Predict on test data
test_mse = mean_squared_error(y_test, y_test_pred)
print(f"Case 2 - With Validation (Separate Test Set): Mean Squared Error = {test_mse:.4f}")
```

```python
# Explanation
if train_mse < test_mse:
    print("\nThe performance on the training data is better than on the test data")
else:
    print("\nThe performance is consistent, indicating no significant overfitting")
```
