# Evaluate the Elastic Net model on the housing dataset using repeated 10-fold cross-validation. Report the average mean absolute error (MAE) obtained across the three repetitions of the 10-fold cross-validation.

```python
# import packages

from numpy import mean
from numpy import std
from numpy import absolute
from pandas import read_csv
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import ElasticNet
```

### Loads the housing dataset from a public repository. Converts the dataset into a NumPy array. Splits the data into:
- X: Input features (all columns except the last).
- y: Target variable (last column, house price).

```python
# load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv'
dataframe = read_csv(url, header=None)
data = dataframe.values
X, y = data[:, :-1], data[:, -1]
```

### Define the Elastic Net model
- alpha=1.0: Controls the overall strength of regularization.
- l1_ratio=0.5: Balances L1 and L2 penalties equally.
- 0 → Ridge regression
- 1 → Lasso regression

```python
# define model
model = ElasticNet(alpha=1.0, l1_ratio=0.5)
```

### Define the evaluation strategy: 10-fold cross-validation: 
Data is split into 10 folds. Repeated 3 times: Reduces variance in performance estimates. random_state: Ensures reproducibility. Each data point is used for testing 30 times in total.

```python
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
```

### Evaluate the model: Trains and evaluates the model using the defined cross-validation.
- Use Mean Absolute Error (MAE) as the performance metric.
- Scikit-learn returns negative MAE by convention (higher is better).
- n_jobs=-1 uses all available CPU cores for faster computation.

```python
# evaluate model
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
```

### Converts negative MAE values to positive for interpretability.
- Mean MAE across all cross-validation runs.
- Standard deviation, indicating variability in performance.

```python
# force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```
