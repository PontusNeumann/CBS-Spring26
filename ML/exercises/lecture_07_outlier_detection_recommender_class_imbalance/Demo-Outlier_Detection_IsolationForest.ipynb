```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('patient.csv')
```

```python
# Feature selection for outlier detection
features = ["Patient_Spend", "Patient_Visits"]
X = df[features]
```

```python
# Visualization before outlier detection
plt.scatter(X['Patient_Spend'], X['Patient_Visits'], c='blue', label='Data points')
plt.title('Data Distribution Before Outlier Detection')
plt.xlabel('Patient Spend')
plt.ylabel('Patient Visits')
plt.legend()
plt.show()
```

```python
# TODO: Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
df['IsoForest_Outlier'] = iso_forest.fit_predict(X)
```

```python
# # TODO: Marking outliers: -1 indicates outliers in both methods
df['IsoForest_Outlier_Flag'] = df['IsoForest_Outlier'] == -1
```

```python
# # TODO: Visualizing the results
outliers_iso = df[df['IsoForest_Outlier_Flag']]
```

```python
# # TODO: Isolation Forest Outlier Visualization
plt.scatter(X['Patient_Spend'], X['Patient_Visits'], c='blue', label='Normal Data')
plt.scatter(outliers_iso['Patient_Spend'], outliers_iso['Patient_Visits'], c='red', label='IsoForest Outliers')
plt.title('Isolation Forest Outlier Detection')
plt.xlabel('Patient Spend')
plt.ylabel('Patient Visits')
plt.legend()
plt.show()
```

```python
# Summary
print("Outliers detected by Isolation Forest:", outliers_iso.shape[0])
```
